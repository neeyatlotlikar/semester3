```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import girvan_newman
```

```python
# Create a sample graph
G = nx.karate_club_graph()  # benchmark social network
```

```python
# Compute communities using the Girvan-Newman algorithm
communities_generator = girvan_newman(G)
top_level_communities = next(communities_generator)
community_list = [list(c) for c in top_level_communities]

# Print community results
print("Communities detected:")
for idx, community in enumerate(community_list):
    print(f"Community {idx + 1}: {community}")
```

```python
# Visualize communities with color groups
colors = ['r', 'g', 'b', 'y']
node_colors = [None] * G.number_of_nodes()
for idx, community in enumerate(community_list):
    for node in community:
        node_colors[node] = colors[idx % len(colors)]

nx.draw(G, node_color=node_colors, with_labels=True)
plt.title("Community structure detected by Girvan-Newman")
plt.show()
```
